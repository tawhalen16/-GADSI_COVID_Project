As with states, many municipalities within states have had varying responses to the Covid-19 pandemic. As a test case, we chose New York State to assess if policy differences between state regions had an association with twitter sentiment scores.

```python
import pandas as pd
import numpy as np
import datetime
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 60)
```

```python
tweets = pd.read_csv('./combined_us.csv')
```

```python
top_cities_nyc = (tweets['state'] == 'NY')
top_cities_nyc = tweets[top_cities_nyc]
# top_cities_solo = top_cities_nyc[['city', 'state']]
```

```python
top_cities_nyc['city'].value_counts()[0:20]
```

```python
top_cities_nyc['city'].replace(['Manhattan', 'Brooklyn', 'Queens',                     
                                'Bronx' , 'Staten Island' ], 'New York City', inplace = True)
```

```python
top_cities_nyc
```

The boroughs of New York City were collapsed into one 'New York City' label because NYC is a very centralized city despite its unique borough identities. The mayors office and city council make all important policy decisions that are not superseded by state authority.

```python
top_cities_nyc['city'].value_counts()[0:20]
```

The top cities in New York state for tweets are NYC, Croton-on-Hudson, Buffalo, Yonkers, Syracuse, North Valley Stream, New Rochelle, Saratoga Springs, Delmar. There is notably a massive difference between New York City and any other city in the state. 

```python
city_policy = pd.read_excel('./city_policies.xlsx')
```

```python
city_policy
```

City policies were collected from a combination of city websites, press releases, and local newspapers. A full list of sources can be found [here](https://docs.google.com/document/d/1vCinvwVOlSMY2jlwrJqPWo1YVEa3VZivqq1aydYlVcY/edit?usp=sharing)

```python
city_policy = top_cities_nyc.merge(city_policy, on = 'city')
```

```python
city_policy
```

```python
city_policy.to_csv('city_policies_ny.csv', index= False)
```

After combining the policies with the relevant tweets, the data is ready for sentiment analysis and mapping.
